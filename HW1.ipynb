Была выбрана цветовая обработка, а именно перевод из RGB в YUV (цветовое пространство в котором Y отвечает за интенсивность, а UV за цвет). И уже в нем уменьшить интенсивность, потом сново перевести в RGB

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
from skimage.color import rgb2yuv, yuv2rgb
```

```python
img = imageio.imread('dog.jpg')
```

```python
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
img = img.astype(np.float32) / 255
```

```python
img_copy = np.copy(img)
```

Перевод в YUV, изменение интенсивности (деление Y канала на 2), обратный перевод и вывод результата

```python
img_yuv = rgb2yuv(img_copy) 
```

```python
img_yuv[:, :, 0] /= 2
img_b = yuv2rgb(img_yuv) 
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_b)
```
